```python
# Create a dictionary to map the current values in 'senioridade' to Portuguese
senioridade_translation = {
    'SE': 'Sênior',
    'MI': 'Pleno',
    'EN': 'Júnior',
    'EX': 'Executivo'
}

# Replace the values in the 'senioridade' column
df['senioridade'].replace(senioridade_translation, inplace=True)

# Display the value counts to verify the changes
display(df['senioridade'].value_counts())
```

```python
df.rename(columns={
    'work_year': 'ano',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'remoto',
    'company_location': 'empresa',
    'company_size': 'tamanho_empresa'
}, inplace=True)

display(df.columns)
```

```python
# Create a dictionary to map the current values in 'contrato' to Portuguese
contrato_translation = {
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Tempo Parcial',
    'FL': 'Freelance'
}

# Replace the values in the 'contrato' column
df['contrato'].replace(contrato_translation, inplace=True)

# Display the value counts to verify the changes
display(df['contrato'].value_counts())
```

```python
# Create a dictionary to map the current values in 'remoto' to Portuguese
remoto_translation = {
    0: 'Presencial',
    50: 'Híbrido',
    100: 'Remoto'
}

# Replace the values in the 'remoto' column
df['remoto'].replace(remoto_translation, inplace=True)

# Display the value counts to verify the changes
display(df['remoto'].value_counts())
```

```python
# Create a dictionary to map the current values in 'tamanho_empresa' to Portuguese
tamanho_empresa_translation = {
    'S': 'Pequena',
    'M': 'Média',
    'L': 'Grande'
}

# Replace the values in the 'tamanho_empresa' column
df['tamanho_empresa'].replace(tamanho_empresa_translation, inplace=True)

# Display the value counts to verify the changes
display(df['tamanho_empresa'].value_counts())
```

```python
import numpy as np

#criação dataframe de teste para usar de exemplo
df_salarios = pd.DataFrame({
    'nome': ['Ana', 'Bruno', 'Carlos', 'Daniele', 'Val'],
    'salario': [4000, np.nan, 5000, np.nan, 100000]
})

#calcula a media salarial e substitui os valores nulos pela media e arredonda valores
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))
#calcula a mediana salarial e substitui os valores nulos pela mediana
df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())
df_salarios
```

```python
df_temperaturas = pd.DataFrame({
    'dia': ['segunda', 'terça', 'quarta', 'quinta', 'sexta'],
    'temperatura': [30, np.nan, np.nan, 28, 27]
})

df_temperaturas['Preenchido_bfill'] = df_temperaturas['temperatura'].bfill()
df_temperaturas
```

```python
df_cidades = pd.DataFrame({
    'nome': ['Ana', 'Bruno', 'Carlos', 'Daniele', 'Val'],
    'cidade': ['São Paulo', np.nan, 'Curitiba', np.nan, 'Belém']
})

df_cidades['cidade_preenchida'] = df_cidades['cidade'].fillna('Não informado')
display(df_cidades)
```

```python
import pycountry

# Função para converter ISO-2 para ISO-3
def iso2_to_iso3(code):
    try:
        return pycountry.countries.get(alpha_2=code).alpha_3
    except:
        return None

# Criar nova coluna com código ISO-3
df_limpo['residencia_iso3'] = df_limpo['residencia'].apply(iso2_to_iso3)

# Calcular média salarial por país (ISO-3) para Data Scientists
df_ds = df_limpo[df_limpo['cargo'] == 'Cientista de Dados']
media_ds_pais = df_ds.groupby('residencia_iso3')['salario'].mean().reset_index()

# Gerar o mapa
fig = px.choropleth(media_ds_pais,
    locations='residencia_iso3',
    color='salario',
    color_continuous_scale='rdylgn',
    title='Média salarial por país de Cientistas de Dados',
    labels={'usd': 'Salário médio em USD', 'residencia_iso3': 'País'})

fig.show()
```
